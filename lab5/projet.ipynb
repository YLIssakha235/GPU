- Sortie:

On doit implementer trois types de ressorts 

1. Structural
    - voisins horizontaux / verticaux

2. shear 
    - diagonales

3. Bending
    - voisins à distance 2


chaque ressort stocke 
- index sommet A
- index sommet B
- longueur au repos

## Etape 5 : Compute Shader (version minimale)

on commence simple, sans la collision

Dans le shader: 
1. lire position + vitesse 
2. appliquer gravité 
3. sommer forces des ressorts (Hooke)
4. amortissement
5. intégrer vitesse + position

Formule ressort: F = −k(∣x∣ − L0​)x^
