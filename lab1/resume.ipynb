# CPU & GPU

1) CPU : Central Processing Unit

c'est un processeur optimisé pour faire des tâches différentes, une par une, mais très bien.

Le CPU est :
- Flexible 
- Bon pour la logique complexe
- Exécute peu de threads (fils) mais très intelligents
- parfait pour organiser le programme


Le CPU = chef d'orchestre : il prépare, configure, gère.


2) GPU : Machine de calcul massif en parallèle 

c'est le Graphics Processing Unit, optimisé pour faire des milliers de calculs IDENTIQUES en même temps.

Il est :

- Très bon pour la parallélisation
- Mille fois plus rapide que le CPU pour des calculs répétitifs
- moins flexique que le CPU

CPU: cerveau qui organise
GPU: calcul parallèle massif






# Shader

Un shader, c'est un "mini-programme" qui s'exécute sur le GPU. 
il existe plusieurs types (WebGPU, Vulkan, DirectX)

- Types de shaders importants:

    1) Vertex Shader
    Transforme la position des vertices:
    -   passe de coordonnées locales - monde - caméra - écran
    - applique les matrices (model, view, projection)

    2) Fragment Shader
    Calcul la couleur de chaque pixel
    - Textures
    - lumières 
    - ombres

    3) Compute Shader (celui ::)
    programme libre qui n'est pas lié au rendu 

    on l'utilises pour :

    - calculer les forces des ressorts
    - mettre à jour vitesse/position
    - gérer collision et friction

# Buffer 

c'est une zone memore stockée sur le GPU.

On y met:
- positions des vertices
- vitesses
- ressorts (indices : quel point attaché à quel point)
- paramètres (masses, ks, kc, g, dt, ...)

le CPU remplit les buffers, puis les envoie au GPU. 

le compute shader lit et ecrit dans ces buffers

| Élément             | Rôle                                      |
| ------------------- | ----------------------------------------- |
| **CPU**             | Organise → prépare → envoie données       |
| **GPU**             | Calcule → beaucoup → en parallèle         |
| **Shader**          | Programme GPU (vertex, fragment, compute) |
| **Pipeline**        | Ordre des étapes GPU                      |
| **Buffer**          | Mémoire GPU contenant données             |
| **Compute Shader**  | Mise à jour du tissu (physique)           |
| **Render Pipeline** | Affichage du tissu                        |
