- adapter.request_device_sync(): c'est à dire maintenant donne moi un device GPU réel sur lequel je peux allouer des buffers et lancer mon compute pipeline.


* Deja les 3 etapes importantes :
- recuperer un device
- créer les buffers : (grid_in, grid_out, uniform_params)

CPU :
