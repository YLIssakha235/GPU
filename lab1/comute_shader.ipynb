## buffers (CPU -GPU) :
on crée des buffers pour stocker des données 
il y'a :
- des buffers d’entrée (données initiales)
- des buffers de sortie (résultats calculés par le GPU)

Conversion NumPy → GPU
- tobytes() → transforme un tableau NumPy en octets

- nbytes → taille du buffer en octets

Parce que le GPU ne comprend que des blocs mémoire binaires, pas des objets Python.

## BufferUsage
- COPY_DST : CPU → GPU (envoyer des données)
- COPY_SRC : GPU → CPU (lire les résultats)
- STORAGE: lecture/écriture dans le compute shader

- uniform = petit, lecture seule

- storage = gros, lecture/écriture (indispensable pour une simulation)

lien avec le projet : 
- entrée : positions, vitesses, springs
- sortie : nouvelles positions, nouvelles vitesses

## Compute Shader
- le shader est un programme GPU
- il s’exécute en parallèle
- il
    - lit les buffers
    - calcule
    - écrit les résultats
Chargé depuis un fichier .wgsl
Utilisé pour faire du calcul, pas du rendu

Le compute shader est exécuté sur le GPU et effectue les calculs physiques en parallèle à partir des données stockées dans les buffers.

## Bind Group Layout vs Bind Group
- Bind Group Layout : décrit la forme (types, bindings, read/write) (a quoi vont ressembler les entrées du shader)
- Bind Group : met les vrais buffers derrière ces bindings

Analogie :
- layout = plan de prises
- bind group = les câbles branchés

Sans bind group : le shader ne voit rien

## Compute Pipeline
Le compute pipeline est l’objet qui relie
- le shader WGSL
- le pipeline layout
- le point d'entrée (main)
- les bind groups
- et permet au GPU d’exécuter le calcul.
voici le programme GPU prét a etre executé
(config, aucun calcul)

il fait :
- prend le shader WGSL
- prend le pipeline layout (bind group layouts) (voici les ressourceds accessibles au shader)
- définit le point d’entrée (main)
- dit au GPU : voici comment exécuter ce calcul

Il est créé une seule fois, pas à chaque frame. 

Une frame : est une itération complète du programme GPU (un tour de boucle) ou on calcul parfois on affiche meme s'il n'ya pas d'affichage on parle de frame.
1 frame : 1 pas de simulation 

### la frame (calcul réel)
- command Encoder : enregistre une liste de commandes GPU
- analogie: je note sur une feuille tout ce que le GPU devra faire (il trvaille pas encore)

- Compute  pass : section commande encoder dédiée au compute (maintenant, ce qui est du calcul GPU)

- set_pipeline : on dit (utilise ce compute pipeline)

- set_bin_group: voci les buffers conrect a utiliser
